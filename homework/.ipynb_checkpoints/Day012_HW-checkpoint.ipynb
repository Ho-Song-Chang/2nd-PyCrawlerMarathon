### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

root = BeautifulSoup(r.text, "html5lib")
for d in root.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
category = {}
for data in root.find(class_="part_list_2").find_all('h3'):
    title = data.find('a').text
    date = data.find(class_ = 'date').text
    tag = data.find('em').text
    category.setdefault(tag,[])
    category[tag].append({
        'title': title,
        'date': date
    })
category
```

### ③ 哪一個類別的文章最多

```python
data = []
for i in category:
    print(i, len(category[i]))
    d = {}
    d['tag'] = i
    d['count'] = len(category[i])
    data.append(d)

for d in sorted(data, key=lambda d: d['count'])[::-1]:
    print(d)
```
